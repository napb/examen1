```python
#Imports
import os
import pandas as pd
import numpy as np
from sklearn import linear_model
import statsmodels.api as sm
```

```python
gasolina = pd.read_csv(os.getcwd() + "/datasets/gasolina/gasolina.csv", index_col=None, header=0, error_bad_lines=False)
gasolina.head()
```

```python
columns = ['Y','Pnc','Puc','Ppt','Pd','Pn','Ps','Pop']
xS = gasolina.loc[:,columns]
y = (gasolina['G'] / gasolina['Pg'])

xS = sm.add_constant(xS)

sm.OLS(y,xS).fit().summary()

#2.A
#Segun se aprecia, el modelo seria:
#Y=308.2542+(0.0364Y)+(-85.9791Pnc)+(-41.9499Puc)+(43.3476Ppt)+(283.5379Pd)+(-765.6020Pn)+(326.0736Ps)+(-0.7863Pop)
#Segun se observa en el R-Cuadrado, el nivel de significancia del modelo es bastante alto (0.990), es decir, 
#casi un modelo estadistico lineal ajustado explicando la variabilidad

#2.B
#Cabe mencionar que las variables, segun este resumen obtenido a traves de la prediccion del 
#modelo con el metodo de minimos cuadrados ordinarios, se deberian dejar todas las variables independientes ya que
#todas tienen un P-Value menor a 0.05
```

```python
#2.C
columns = ['Pnc','Puc']
xS = gasolina.loc[:,columns]
sm.OLS((gasolina['G'] / gasolina['Pg']), sm.add_constant(xS)).fit().summary()

#Segun se puede apreciar en la estadistica F, el P-Value el valor es practicamente 0, de esta forma el modelo
#estadistico entre Pnc y Puc con la variable estimada G/Pg seria significativo. Pero tambien Puc, no es
#estadisticamente significativa, con un P-Value de 0.167
```
